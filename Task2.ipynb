```python
image = cv2.imread("nun.jpg")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
#blurred = cv2.GaussianBlur(gray, (5, 5), 0)
#blurred = cv2.GaussianBlur(gray, (7, 7), 0)
#blurred = cv2.GaussianBlur(gray, (57, 57), 0)

thresh = cv2.threshold(blurred, 90, 250, cv2.THRESH_BINARY)[1]
#thresh =  cv2.bitwise_not(thresh)
```

```python
cv2.imshow("Image",thresh)
cv2.waitKey(0)
```

```python
# find contours in the thresholded image
cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
```

```python
# loop over the contours
def shapedetect( count ):                    #give number as count of angles)
    for c in cnts:
        M = cv2.moments(c)
        peri = cv2.arcLength(c, True)
        approx = cv2.approxPolyDP(c, 0.04 * peri, True)
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])
        if len(approx) == count and count == 3:
            shape = "triangle"
            cv2.drawContours(image, [c], -1, (0, 255, 0), 2)
            cv2.circle(image, (cX, cY), 7, (255, 255, 255), -1)
            cv2.putText(image, shape, (cX - 20, cY - 20),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
        
        elif len(approx) == count and count == 4:
            cv2.drawContours(image, [c], -1, (0, 255, 0), 2)
            cv2.circle(image, (cX, cY), 7, (255, 255, 255), -1)
            cv2.putText(image, shape, (cX - 20, cY - 20),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
            (x, y, w, h) = cv2.boundingRect(approx)
            ar = w / float(h)
            shape = "square" if ar >= 0.95 and ar <= 1.05 else "rectangle"
         
        elif len(approx) == count and count == 5:
            shape = "pentagon"
            cv2.drawContours(image, [c], -1, (0, 255, 0), 2)
            cv2.circle(image, (cX, cY), 7, (255, 255, 255), -1)
            cv2.putText(image, shape, (cX - 20, cY - 20),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
        elif count == 0:
            shape = "circle"
            print("Hi")
            cv2.drawContours(image, [c], -1, (0, 255, 0), 2)
            cv2.circle(image, (cX, cY), 7, (255, 255, 255), -1)
            cv2.putText(image, shape, (cX - 20, cY - 20),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
    cv2.imshow("Image", image)
    cv2.waitKey(0)
```
