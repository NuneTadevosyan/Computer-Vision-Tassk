To find objects of given color I iterate over pixels and add information of the pixel (color, coordinates) to the dictonary if its color is the given one. 
To separate objects from the set I use DBSCAN algorithm and can count area of each class by comparing pixel count.

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.cluster import DBSCAN
import time
import imageio as iio
from numpy import uint8
import copy
from PIL import Image
import cv2
```

```python
def encode(img_n,r,g,b):
    ret = {}
    _r = []
    _g = []
    _b = []
    _c = []
    for h in tqdm(range(img_n.shape[0])):
        for w in range(img_n.shape[1]):
            if img_n[h][w][0] == r and img_n[h][w][1] == g and img_n[h][w][2] == b:
                _r.append(img_n[h][w][0])
                _g.append(img_n[h][w][1])
                _b.append(img_n[h][w][2])
                _c.append([w, h])
    ret['R'] = _r
    ret['G'] = _g
    ret['B'] = _b
    ret['coordinate'] = _c
    return ret
```
